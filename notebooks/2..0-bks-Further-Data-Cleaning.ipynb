Going to try and use Principal Component Analysis to focus on the
words with the most variation amongst players. Ideally this large
variation could be used for ease of classifying a piece of text to
a player

```python
df = pd.read_csv('../data/processed/Word_Counts.csv')
```

Group the words and plot to see what words occur enough to be worth removing from the dataframe. Words that occur too frequently can be determined to be relatively commonly used english words like "the" or "A". Such commonly used words would provide relatively little variation to the characters.

```python
df.groupby('PlayerLine').sum().plot()
```

```python
outliers = df.groupby('PlayerLine').sum() > 1000
```

```python
outliers = outliers.reset_index()
```

```python
outliers = outliers[outliers['PlayerLine.1']!=True]
```

```python
df = df[df.PlayerLine.isin(outliers.PlayerLine.tolist())]
```

```python
df.plot()
```
